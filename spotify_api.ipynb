#### Unificando funciones

```python
def get_token(client_id, client_secret):
    encoded = base64.b64encode(bytes(client_id+':'+client_secret, 'utf-8'))
    params = {'grant_type':'client_credentials'}
    header = {'Authorization': 'Basic' +' '+ str(encoded,'utf-8')}
    r = requests.post('https://accounts.spotify.com/api/token', headers = header, data= params)
    if r.status_code !=200:
        print('Error en la request.', r.json())
        return None
    print('Token valido por {} segundos.'.format(r.json()['expires_in']))
    return r.json()['access_token']
```

### Funcion obtener informaci√≥n completa

```python
def obtener_discografia(artist_id, token, return_name = False, page_limit = 50, country= None):
    url = url_base + endpoint_artist_album
    header = {'Authorization': f'Bearer {token}'}
    params = {'limit': page_limit,
              'offset': 0,
              'country': country}
    lista = []
    r = requests.get(url, params=params, headers=header)

    if r.status_code != 200:
        print('Error en la request.', r.json())
        return None

    if return_name:
        lista += [(item['id'], item['name']) for item in r.json()['items']]
    else:
        lista += [item['id'] for item in r.json()['items']]

    while r.json()['next']:
        r = requests.get(r.json()['next'], headers=header)
        if r.status_code != 200:
            print('Error en la request.', r.json())
            return None

        if return_name:
            lista += [(item['id'], item['name']) for item in r.json()['items']]
        else:
            lista += [item['id'] for item in r.json()['items']]
    
    return lista
```

#### Obtener todas las canciones del artista

```python
def obtener_tracks(artist_id, token, return_name = False, page_limit = 50, country= None):
    url = url_base + endpoint_album_tracks
    header = {'Authorization': f'Bearer {token}'}
    params = {'limit': page_limit,
              'offset': 0,
              'country': country}
    lista = []
    r = requests.get(url, params=params, headers=header)

    if r.status_code != 200:
        print('Error en la request.', r.json())
        return None

    if return_name:
        lista += [(item['id'], item['name']) for item in r.json()['items']]
    else:
        lista += [item['id'] for item in r.json()['items']]

    while r.json()['next']:
        r = requests.get(r.json()['next'], headers=header)
        if r.status_code != 200:
            print('Error en la request.', r.json())
            return None

        if return_name:
            lista += [(item['id'], item['name']) for item in r.json()['items']]
        else:
            lista += [item['id'] for item in r.json()['items']]
    
    return lista
```

```python
for album in obtener_discografia(kygo_id, token, return_name=True, country='CL'):
    print(album[1])
    for track in obtener_tracks(album[0], token, return_name=True, market='CL'):
        print('\t', track[1])
```
